```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dat = pd.read_csv("iris.csv")
dat
```

```python
dat.describe()
```

```python
dat.hist("sepal.length", bins=30)
plt.title("Histogram")
plt.xlabel("Width")
plt.ylabel("Counts")
```

```python
plt.figure(figsize=(6, 10))

dat.hist("sepal.length", bins = 20)
plt.title("Wahthef")
plt.xlabel("xxxxx")
plt.ylabel("yyyyyy")
```

```python
# Set figure size 
plt.figure(figsize=(10,8)) # width 10 inch, height 8 inch

# Histogram of Sepal Length
plt.subplot(2, 2, 1) # 1st subplot position in a (2,2) grid
plt.hist(dat["sepal.length"],bins=30, color='lightblue')
plt.title('Subplot 1')

# Histogram of Sepal Width
plt.subplot(2, 2, 2) # 2nd subplot position in a (2,2) grid
plt.hist(dat["sepal.width"],bins=30, color='lightblue')
plt.title('Subplot 2')

# Histogram of Petal Length
plt.subplot(2, 2, 3)
plt.hist(dat["petal.length"],bins=30, color='lightblue')
plt.title('Subplot 3')

# Histogram of Petal Width
plt.subplot(2, 2, 4)
plt.hist(dat["petal.width"],bins=30, color='lightblue')
plt.title('Subplot 4')

plt.tight_layout()
plt.show()
```
